```python
# drop all columns which have vehiclegroup in the name except vehiclegroup01
def drop(data):
    return data.drop(data.columns[data.columns.str.contains('vehiclegroup')], axis=1)
```

## SARIMAX model

```python
df_a = countryA(data)
df_a.head(20)
```

```python
# create data frame with month and sales_actuals_monthly__vehiclegroup01__orderintake
df_a01 = df_a[['month','sales_actuals_monthly__vehiclegroup01__orderintake']]

# df_a01 als time series
df_a01 = df_a01.set_index('month')
df_a01.head(20)
```

```python
#  plot time series
plt.plot(df_a01)
plt.show()
```

### Trend?

```python
# delete rows after month 258
df_a01 = df_a01[df_a01.index <= 258]
```

```python
# missing values in time series?
df_a01.isnull().sum()
```

```python
# dickey fuller test
from statsmodels.tsa.stattools import adfuller

adf = adfuller(df_a01['sales_actuals_monthly__vehiclegroup01__orderintake'])
print('ADF Statistic: %f' % adf[0])
print('p-value: %f' % adf[1])
# hypotheses
# H0: time series is not stationary
# H1: time series is stationary
# p-value > 0.05 -> H0 cannot be rejected -> time series is not stationary
# here: time series is stationary
```

```python
# kpss test for stationarity
from statsmodels.tsa.stattools import kpss

kpsstest = kpss(df_a01['sales_actuals_monthly__vehiclegroup01__orderintake'], regression='c')
print('KPSS Statistic: %f' % kpsstest[0])
print('p-value: %f' % kpsstest[1])
# hypotheses
# H0: time series is stationary
# H1: time series is not stationary
# here: time series is stationary
```

### Seasonality

```python
# acf and pacf
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf

plot_acf(df_a01)
plt.show()

plot_pacf(df_a01)
plt.show()
```

--> yearly seasonality
-> diff(12)

```python
# differencing with lag 12
df_a01_diff = df_a01.diff(periods=12)
```

```python
# plot differenced time series
plt.plot(df_a01_diff)
```

```python
# acf and pacf of differenced time series
plot_acf(df_a01_diff)
plt.show()
```
