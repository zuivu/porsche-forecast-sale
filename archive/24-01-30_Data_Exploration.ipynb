```python
# Python 
train = vec1_A_intern
train["y"] = train["sales_actuals_monthly__vehiclegroup01__orderintake"]
train = train.drop("sales_actuals_monthly__vehiclegroup01__orderintake", axis = 1)

train["ds"] = train.index

train.head()
```

```python
# Convert the 'ds' column to datetime format
start_date = '2000-01-01'  # Adjust this date to your needs
train['ds'] = pd.to_datetime(train['ds'].apply(lambda x: pd.DateOffset(months=x) + pd.Timestamp(start_date)))
```

```python
train.head()
```

```python
from prophet import Prophet

m = Prophet()
m.fit(train)

future = m.make_future_dataframe(periods=18, freq = "MS")
future.tail()

forecast = m.predict(future)
forecast.tail()
```

```python
fig1 = m.plot(forecast)
```

```python
train_pro = train.iloc[:240, :]
test_pro = train.iloc[240:, :]
```

```python
test_pro.shape
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

m = Prophet()

# Fit the model with train set
m.fit(train_pro)

future = m.make_future_dataframe(periods=18, freq = "MS")

# Make predictions on the test set dates
forecast = m.predict(future)

forecast = forecast.iloc[240:, :]

# Calculate errors
mse = mean_squared_error(test_pro['y'], forecast['yhat'])
mae = mean_absolute_error(test_pro['y'], forecast['yhat'])

print('Mean Squared Error:', mse)
print('Mean Absolute Error:', mae)

```

```python


# Create a new figure
plt.figure()

# Plot y_test
plt.plot(test_pro["y"] , label='Actual')

# Plot y_pred
plt.plot(forecast["yhat"], label='Predicted')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
