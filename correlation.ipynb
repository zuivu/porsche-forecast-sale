```python
import pandas as pd
```

```python
df = pd.read_csv("Data/kitCapstoneJan24_data.csv", delimiter=";")
df.head()
```

```python
col_names = list(df.columns)
print(col_names[2:10])
```

```python
vehicle_list = []
sales_info_list  = []

for col in col_names:
    # vehicle group 
    if "vehiclegroup" in col:
        vehicle_name = col.split("__")[1]
        if vehicle_name not in vehicle_list: 
            vehicle_list.append(vehicle_name)

    # sales
    if "sales" in col:
        col_info = col.split("__")
        try:
            sale_info = col_info[0] + "__" + col_info[2]
            if sale_info not in sales_info_list: 
                sales_info_list.append(sale_info)
        except IndexError:
            print(col_info)
            break
```
